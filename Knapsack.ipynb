```python
import pandas as pd
from time import *
plastic_data=pd.read_csv("knapsack1.csv")
```

```python
plastic_data
```

```python
#0-1 Knapsack
wt=(plastic_data['weight']).to_numpy()
wt
```

```python
pf=(plastic_data['profit']).to_numpy()
pf
```

```python
#Setting knapsack capacity to 500
n=len(wt)
W=500
start = time()
F = [[0 for i in range(0,n)] for j in range(0,W)] 
for i in range(1,n):
    for j in range(1,W):
            if (j-wt[i])<0:
                F[i][j]=F[i-1][j]
            elif (j-wt[i])>=0:
                F[i][j]=max(F[i-1][j],pf[i]+F[i-1][j-wt[i]])

stop = time()
print("Execution time of 0-1 knapsack is {} seconds.".format(stop-start))
print('Maximum Profit is:',F[n-1][W-1])
```

```python
#Fractional Knapsack
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv("knapsack1.csv")
data['Ratio']=data['profit']/data['weight']
data
```

```python
data=data.sort_values('Ratio',ascending=False)
data
```

```python
data['Solution set']=[0 for i in range(0,len(data))]
data
```

```python
#Setting capacity of knapsack to 500
W=500
total=profit=0
for i in range(0,len(data)):
    if (total+data['weight'].iloc[i] <= W):
        data['Solution set'].iloc[i]=1
        total=total+data['weight'].iloc[i]
    else:
        data['Solution set'].iloc[i]=(W-total)/data['weight'].iloc[i]
        total = W
        break
print(data)
```
